```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

df1 = pd.read_csv("가격_구간별_총_주문수량.csv")

plt.figure(figsize=(8, 5))
sns.barplot(x="가격 구간", y="주문 수량", data=df1, palette="YlGnBu")
plt.title("가격 구간별 총 주문 수량", fontsize=14, fontweight='bold')
plt.xlabel("가격 구간", fontsize=12)
plt.ylabel("주문 수량", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 해석
가격이 높을수록 주문량이 줄어들기는 하나, 고가 피자의 주문량도 매우 높은 수준

중가 메뉴가 가장 많은 주문량을 기록하며 중심축 역할을 함

저가와 고가는 수요가 명확히 갈리지만, 고가도 수요가 안정적으로 존재함

### 결론
중가 피자가 전체 판매의 중심 → 주력 제품군으로 포지셔닝

고가 메뉴는 수요가 꺾이지 않으므로 프리미엄 라인업으로 고정 고객을 유도

저가 메뉴는 가격 민감층이나 점심 메뉴/테이크아웃 메뉴로 활용 가능

```python
df2 = pd.read_csv("재료수_별_총_주문수량.csv")

plt.figure(figsize=(8, 5))
sns.barplot(x="재료 수", y="주문 수량", data=df2, palette="Blues_d")
plt.title("재료 수 별 총 주문 수량", fontsize=14, fontweight='bold')
plt.xlabel("재료 수", fontsize=12)
plt.ylabel("주문 수량", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 해석
고객은 재료가 너무 적은 피자를 비선호하는 경향

하지만 재료가 너무 많아지는 (7~8개) 경우 주문량 증가가 정체되거나 감소하는 모습도 존재

시각적으로 5~6개의 구간이 가장 뚜렷하게 높고 안정적임

### 결론

재료는 5~6개가 '맛있어 보이면서도 부담 없는' 최적 조합

2~3개짜리 단순 메뉴는 피자 본연의 매력을 줄일 수 있음

7개 이상 재료는 관리 비용, 제조 복잡성도 증가 → 비효율적일 수 있음

```python
df3 = pd.read_csv("토핑_종류별_등장횟수.csv")
top10 = df3.head(10)

plt.figure(figsize=(8, 6))
sns.barplot(y="토핑", x="등장 횟수", data=top10, palette="viridis")
plt.title("상위 10개 토핑 등장 횟수", fontsize=14, fontweight='bold')
plt.xlabel("등장 횟수", fontsize=12)
plt.ylabel("토핑", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

### 해석
이 재료들은 고객의 기본적이고 보편적인 입맛에 부합

시각적으로 크기(빈도 수)가 가장 크며, 이는 해당 토핑이 핵심 재료임을 뜻함

### 분석
Garlic, Tomatoes 등은 “모든 메뉴에서 기본 베이스로 사용해도 괜찮은 재료”

피자 신메뉴 개발 시, 이 재료를 중심으로 구성하거나 변주하는 전략이 안정적

현지화 시에도 이 재료들은 보편적으로 수용 가능한 식재료

```python
df4 = pd.read_csv("피자_별_단가_및_총_주문수량.csv")

plt.figure(figsize=(7, 5))
sns.scatterplot(x="단가", y="주문 수량", data=df4, color="tomato", s=80, edgecolor="k")
plt.title("피자 단가 vs 주문 수량", fontsize=14, fontweight='bold')
plt.xlabel("단가 ($)", fontsize=12)
plt.ylabel("주문 수량", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 해석
가격이 높더라도 구매 장벽으로 작용하지 않음 → 고객은 “가성비”보다 “가치”에 반응

인기 메뉴는 단가가 높더라도 맛·재료 구성 등으로 납득이 가능함

### 결론

고가 메뉴는 고객에게도 매력적인 제품일 수 있음

고가이더라도 맛/재료 조합/브랜드 신뢰만 확보되면 충분히 구매 유도 가능

“고가 = 비인기”라는 통념을 버리고 프리미엄 상품을 전략적으로 운영할 것


```python
# ✅ 데이터 불러오기
df = pd.read_csv("피자_별_재료_수.csv")

# ✅ 2~8까지 x축 범위 고정
x_range = list(range(2, 9))
value_counts_full = df["재료 수"].value_counts().reindex(x_range, fill_value=0)

# ✅ 색상 팔레트 생성
colors = sns.color_palette("Spectral", n_colors=len(x_range))

# ✅ 시각화
plt.figure(figsize=(10, 6))
bars = plt.bar(x=x_range, height=value_counts_full.values, color=colors, edgecolor='black')

# ✅ 눈금 수동 지정
plt.xticks(ticks=x_range)

# ✅ 그래프 꾸미기
plt.title("피자별 재료 수 분포 (2~8)", fontsize=16, weight='bold')
plt.xlabel("재료 수", fontsize=12)
plt.ylabel("피자 개수", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

### 해석
인기 피자는 대부분 복합적인 재료 조합을 갖고 있음

고객은 “풍부하고 다양한 맛”을 선호하는 경향이 강함

단순한 구성이 아닌, 조화된 다재료 조합이 미각을 자극

### 결론

신메뉴 기획 시, 재료 수는 5~7개 수준에서 고정하는 것이 이상적

핵심 재료를 잘 조합해 풍성하면서도 단가 관리가 가능한 메뉴 구성이 중요

```python
# CSV 파일 불러오기
df = pd.read_csv("피자_종류별_재료_피자수_요약.csv")

# 피자 종류별 상위 5개 재료 선택
top5_by_category = (
    df.sort_values(["피자 종류", "피자 메뉴 개수"], ascending=[True, False])
    .groupby("피자 종류")
    .head(5)
)

# 카테고리별 시각화
categories = top5_by_category["피자 종류"].unique()

for category in categories:
    cat_data = top5_by_category[top5_by_category["피자 종류"] == category]
    
    plt.figure(figsize=(8, 5))
    sns.barplot(
        data=cat_data,
        x="피자 메뉴 개수",
        y="재료",
        palette="Set2",
        edgecolor="black"
    )
    plt.title(f"{category} 카테고리 - Top 5 재료", fontsize=14, weight='bold')
    plt.xlabel("사용된 피자 수")
    plt.ylabel("재료")
    plt.grid(axis='x', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python

df = pd.read_csv("피자_종류별_재료_피자수_요약.csv")

# 카테고리에서 상위 5개 재료만 추출
top5_by_category = (
    df.sort_values(["피자 종류", "피자 메뉴 개수"], ascending=[True, False])
    .groupby("피자 종류")
    .head(5)
)

# x축 라벨 생성: "카테고리 - 재료"
top5_by_category["카테고리별 재료"] = top5_by_category["피자 종류"] + " - " + top5_by_category["재료"]

# 시각화
plt.figure(figsize=(14, 6))
sns.barplot(
    x="카테고리별 재료",
    y="피자 메뉴 개수",
    hue="피자 종류",
    data=top5_by_category,
    dodge=False,
    palette="Set2"
)
plt.title("피자 카테고리별 Top 5 재료", fontsize=16, weight='bold')
plt.xlabel("카테고리 - 재료", fontsize=12)
plt.ylabel("사용된 피자 수", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

### 해석

| 피자 종류       | 특성      | 재료 활용 특징                                     |
| ----------- | ------- | -------------------------------------------- |
| **Classic** | 전통적인 피자 | 햄, 파인애플, 치즈 등 기본 재료에 집중                      |
| **Supreme** | 푸짐한 구성  | 다양한 재료가 균형 있게 분포됨 (햄, 소시지, 채소 등)             |
| **Veggie**  | 건강 지향   | Mushroom, Onion, Tomato, Spinach 등 야채류 비중 높음 |
| **Chicken** | 단백질 중심  | 치킨 기반에 Mushroom, Onion 등 부재료 다양히 활용          |

### 결론

각 피자 카테고리는 뚜렷한 재료 사용 패턴이 있어, 메뉴 구성을 할 때 명확한 컨셉 설정이 가능합니다.

메뉴판에서 카테고리별 특징을 강조하면 고객이 원하는 취향에 따라 빠르게 선택할 수 있습니다:
