```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기
df = pd.read_csv('Pizza Sales Dataset.csv')

# 컬럼명 한글로 변경
df.rename(columns={
    'pizza_id': '피자 ID',
    'order_id': '주문 ID',
    'pizza_name_id': '피자 이름 ID',
    'quantity': '주문 수량',
    'order_date': '주문 날짜',
    'order_day': '주문 요일',
    'order_time': '주문 시간',
    'unit_price': '단가',
    'total_price': '총액',
    'pizza_size': '피자 사이즈',
    'pizza_category': '피자 종류',
    'pizza_ingredients': '피자 재료 목록',
    'pizza_name': '피자 이름'
}, inplace=True)

# ✅ 요일별 매출 집계
day_sales = df.groupby('주문 요일')['총액'].sum().reset_index()

# ✅ 요일 순서 지정 (정렬을 위해)
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_sales['주문 요일'] = pd.Categorical(day_sales['주문 요일'], categories=weekday_order, ordered=True)
day_sales = day_sales.sort_values('주문 요일')

# ✅ 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x='주문 요일', y='총액', data=day_sales, palette='viridis')

plt.title('요일별 매출 분석')
plt.xlabel('요일')
plt.ylabel('총 매출 (총액)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기 및 한글 컬럼명 적용은 이미 되어 있다고 가정

# 주문 시간 전처리
df['주문 시간'] = pd.to_datetime(df['주문 시간'], format='%H:%M:%S')

# 시간대 정의 함수
def get_time_slot(hour):
    if 6 <= hour < 11:
        return '아침 (6~11시)'
    elif 11 <= hour < 15:
        return '점심 (11~15시)'
    elif 15 <= hour < 18:
        return '오후 (15~18시)'
    elif 18 <= hour < 22:
        return '저녁 (18~22시)'
    else:
        return '심야 (22~익일6시)'

# 시간대 컬럼 생성
df['시간대'] = df['주문 시간'].dt.hour.apply(get_time_slot)

# 요일/시간대 기준 주문 수량 집계
time_by_day = df.groupby(['주문 요일', '시간대'])['주문 수량'].sum().reset_index()

# 요일/시간대 순서 지정
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
time_slot_order = ['아침 (6~11시)', '점심 (11~15시)', '오후 (15~18시)', '저녁 (18~22시)', '심야 (22~익일6시)']
time_by_day['주문 요일'] = pd.Categorical(time_by_day['주문 요일'], categories=weekday_order, ordered=True)
time_by_day['시간대'] = pd.Categorical(time_by_day['시간대'], categories=time_slot_order, ordered=True)

# 정렬
time_by_day = time_by_day.sort_values(['주문 요일', '시간대'])

# 피벗 테이블: 시간대별 x축, 요일별 선분
pivot_df = time_by_day.pivot(index='시간대', columns='주문 요일', values='주문 수량')

# 시각화
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# 꺾은선 그래프 그리기
for day in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[day], marker='o', label=day)

plt.title('요일별 시간대 주문량 추이')
plt.xlabel('시간대')
plt.ylabel('주문 수량')
plt.xticks(rotation=45)
plt.legend(title='요일')
plt.tight_layout()
plt.show()
```

```python
결과 분석 : 업무 일 중 중반 요일의 점심시간대가 활발하게 주문 많음. 금요일과 토요일 저녁 시간대 주문량 급증. 금요일과 토요일에 심야주문량 상대적 많음.
```
